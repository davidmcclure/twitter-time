```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
import numpy as np
```

```python
from twitter_time.models import MinuteCount
```

```python
series = MinuteCount.overall_series()

plt.figure(figsize=(15, 8))

plt.title('Overall word volume')
plt.xlabel('Minute')
plt.ylabel('Word count')

plt.hist(
    np.arange(len(series)),
    bins=len(series),
    weights=series,
    edgecolor='black',
    zorder=3,
)

plt.show()
```
