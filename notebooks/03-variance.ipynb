```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
from twitter_time.models import MinuteCount
```

```python
tokens = MinuteCount.token_counts()
```

```python
data = []

for i, (token, count) in enumerate(tokens.items()):
    
    series = MinuteCount.token_series(token)
    exp_var = series.sum() * (1/100) * (1 - 1/100)
    
    data.append((count, series.var(), exp_var))
    
    if i % 1000 == 0: print(i)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(
    data,
    columns=('count', 'var', 'exp'),
    index=tokens.keys(),
)
```

```python
pd.set_option('display.max_rows', 1000)

df.sort_values('var', ascending=False).head(100)
```

```python
plt.figure(figsize=(10, 8))

plt.title('Variance, linear scales')
plt.xlabel('Word count')
plt.ylabel('Variance')
plt.scatter(df['count'], df['var'], s=3, c='b')

plt.show()
```

```python
plt.figure(figsize=(10, 8))

plt.title('Variance, log-log')
plt.xlabel('Word count')
plt.ylabel('Variance')
plt.xscale('log')
plt.yscale('log')
plt.scatter(df['count'], df['var'], s=3, c='b')

plt.show()
```

```python
plt.figure(figsize=(10, 8))

plt.title('Observed vs. expected variance')
plt.xlabel('Word count')
plt.ylabel('Variance')
plt.xscale('log')
plt.yscale('log')
plt.scatter(df['count'], df['var'], s=3, c='b')
plt.plot(df['count'], df['exp'])

plt.show()
```

```python
import math

df['score'] = df.index.map(
    lambda t: df.loc[t]['var'] / df.loc[t]['exp']
)
```

```python
df.sort_values('score', ascending=False).head(100)
```

```python
import numpy as np
```

```python
def plot_series(token):
    
    series = MinuteCount.token_series(token)
    
    plt.figure(figsize=(4, 2))

    print(token)
    plt.xlabel('Minutes')
    plt.ylabel('Word count')
    
    plt.hist(
        np.arange(len(series)),
        bins=len(series),
        weights=series,
        edgecolor='black',
        zorder=3,
    )
    
    plt.show()
```

```python
for token in df.sort_values('score', ascending=False).head(100).index:
    plot_series(token)
```
