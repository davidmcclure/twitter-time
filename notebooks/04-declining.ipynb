```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
import numpy as np
```

```python
from twitter_time.models import MinuteCount
```

```python
def plot_series(token, title=None):
    
    series = MinuteCount.token_series(token)
    
    plt.figure(figsize=(10, 6))

    plt.title(title or '"{}"'.format(token))
    plt.xlabel('Minutes')
    plt.ylabel('Word count')
    
    plt.hist(
        np.arange(len(series)),
        bins=len(series),
        weights=series,
        edgecolor='black',
        zorder=3,
    )
    
    plt.show()
```

```python
plot_series('tracked')
```
