```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
import pandas as pd
import ujson
```

```python
with open('../twitter_time/char_minute_counts/counts.json') as fh:
    data = [ujson.loads(line) for line in fh]
```

```python
df = pd.DataFrame(data)
```

```python
df = df.sort_values('minute')
```

```python
def plot_series(char):
    
    print(char)
    
    series = df[df['char'] == char]
    
    plt.figure(figsize=(6, 4))
    plt.xlabel('Minutes')
    plt.ylabel('Char count')
    
    plt.hist(
        series['minute'],
        bins=len(series),
        weights=series['count'],
        edgecolor='black',
        zorder=3,
    )
    
    plt.show()
```

```python
for char in df['char'].unique():
    plot_series(char)
```
