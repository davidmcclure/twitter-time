```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
import numpy as np

from wordfreq import top_n_list

from twitter_time.token_minute_counts.db import TokenMinuteCount
```

```python
def plot_series(token, w=10, h=6):
    
    print(token)
    
    series = TokenMinuteCount.token_series(token)
    
    plt.figure(figsize=(w, h))

    plt.title(token)
    plt.xlabel('Minutes')
    plt.ylabel('Word count')
    
    plt.hist(
        np.arange(len(series)),
        bins=len(series),
        weights=series,
        edgecolor='black',
        
        zorder=3,
    )
    
    plt.show()
```

```python
for token in top_n_list('en', 100):
    plot_series(token, 5, 3)
```
